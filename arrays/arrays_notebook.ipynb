### Remove Duplicates From Sorted Array

`Input: nums = [1,1,2,3,4]`

`Output: [1,2,3,4]`

```python
nums=[1,1,2,3,4]
```

```python
l = 1
for r in range(1, len(nums)):
    if nums[r] != nums[r-1]:
        nums[l] = nums[r]
        l += 1
        
print(nums[:l])
```

```python
nums
```

```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        l = 1
        for r in range(1,len(nums)):
            if nums[r] != nums[r-1]:
                nums[l] = nums[r]
                l +=1


        return nums[:l]
```

```python
s = Solution()
```

```python
s.removeDuplicates(nums)
```

***Basic Array Problems***

*1. Find the maximum element in an array.*

```python
def get_max_value(nums: List[int]) -> int:
    max_value = 0
    for num in nums:
        if num > max_value:
            max_value = num
    return max_value
```

```python
nums=[2,10,10,30,30,30]
```

```python
print(get_max_value(nums))
```

```python
from typing import List

def get_max_value(nums: List[int]) -> int:
    if not nums:
        raise ValueError("Array is empty")

    max_value = nums[0]
    for num in nums[1:]:
        if num > max_value:
            max_value = num
    return max_value
```

```python
nums = [7, -3, 10, 2]
```

```python
print(get_max_value(nums))
```

```python
from typing import List
def get_min_value(nums : List[int]) -> int:
    if not nums:
        raise ValueError("Array is empty")
    
    min_value = nums[0]
    for num in nums[1:]:
        if num < min_value:
            min_value = num
    return min_value
    
```

```python
print(get_min_value(nums))
```
